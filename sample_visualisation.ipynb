```python
import matplotlib.pyplot as plt

# Subplots
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(titanic_df.age, titanic_df.fare)
axs[0, 0].set_title('Subplot 1')

axs[0, 1].scatter(titanic_df.age, titanic_df.fare)
axs[0, 1].set_title('Subplot 2')

axs[1, 0].bar(titanic_df.age, titanic_df.fare)
axs[1, 0].set_title('Subplot 3')

axs[1, 1].hist(titanic_df.age)
axs[1, 1].set_title('Titanic age distribution')

plt.tight_layout()
plt.show()
```
